Import pycaret and data.

```python
import pandas as pd
import pycaret
```

```python
data = pd.read_csv("healthcare_dataset.csv")
data.head()
```

```python
data.columns
```

```python
data['Name']
```

```python
data = data.drop(columns=['Name', 'Doctor', 'Hospital', 'Insurance Provider', 'Billing Amount', 'Room Number', 'Date of Admission', 'Discharge Date'])
```

```python
data.head()
```

```python
from pycaret.classification import *
s = setup(data, target = 'Test Results')
```

```python
best = compare_models(sort = 'AUC')
```

```python
tune_model(best)
```

```python
print(best)
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
new_data = data.copy().drop('Test Results', axis=1)
predictions = predict_model(best, new_data)
predictions.head(50)
```
